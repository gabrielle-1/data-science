### Conceitos Básicos Estatísica em Ciência de Dados

Requirements: 
numpy
matplotlib
seaborn

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = [
        3,
        4,
        5,
        6,
        7,
        7,
        10
]
mid = np.mean(data)
x = range(0, len(data))
plt.scatter(x, data)
plt.axhline(y= mid, color = 'r', linestyle= '-')
plt.show()
```

```python
data = [
        3,
        4,
        5,
        6,
        7,
        7,
        10
]

mid = np.mean(data)
median = np.median(data)
variance = np.var(data)
standard_deviation = np.std(data)
print("Média: ", mid)
print("Mediana: ", median)
print("Variância: ", variance)
print("Desvio Padrão: ", standard_deviation)
```

```python
ax = sns.histplot(data, stat="density", kde=True, color="#74acd4", lw=1)
```
